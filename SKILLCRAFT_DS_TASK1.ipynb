# **BAR PLOT AND HISTOGRAMS**
_________

```python
# Bar chart for gender distribution
gender_counts = df['gender'].value_counts()
```

```python
plt.figure(figsize=(6,4))
gender_counts.plot(kind='bar', color=['red','green'])
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Separate math scores by gender
male_scores = df[df['gender'] == 'male']['math score']
female_scores = df[df['gender'] == 'female']['math score']
```

```python
# Plot overlay histograms
plt.figure(figsize=(8,5))
plt.hist(male_scores, bins=10, alpha=0.5, label='Male', color='purple', edgecolor='black')
plt.hist(female_scores, bins=10, alpha=0.5, label='Female', color='pink', edgecolor='black')

plt.title('Math Score Distribution by Gender')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df['parental level of education'].value_counts().plot(
    kind='bar', figsize=(8,4), color='teal', edgecolor='black'
)
plt.title('Parental Level of Education Distribution')
plt.xlabel('Education Level')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['race/ethnicity'].value_counts().plot(
    kind='bar', figsize=(8,4), color='blue', edgecolor='black'
)
plt.title('race/Ethnicity Distribution')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# --- 2. Test preparation course completion counts ---
plt.figure(figsize=(6,4))
df['test preparation course'].value_counts().plot(
    kind='bar', color=['mediumseagreen','orange'], edgecolor='black'
)
plt.title('Test Preparation Course Completion Counts')
plt.xlabel('Test Preparation Course')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# --- 3. Lunch type distribution ---
plt.figure(figsize=(6,4))
df['lunch'].value_counts().plot(
    kind='bar', color=['orchid','gold'], edgecolor='black'
)
plt.title('Lunch Type Distribution')
plt.xlabel('Lunch Type')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Example: Histogram for math score
plt.figure(figsize=(6,4))
plt.hist(df['math score'], bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

```python
avg_math_by_gender = df.groupby('gender')['math score'].mean()

plt.figure(figsize=(6,4))
avg_math_by_gender.plot(kind='bar', color=['skyblue','lightcoral'], edgecolor='black')
plt.title('Average Math Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Math Score')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['reading score'], bins=10, color='red', edgecolor='black')
plt.title('Distribution of Reading Scores')
plt.xlabel('Reading Score')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

```python
avg_reading_by_gender = df.groupby('gender')['reading score'].mean()

plt.figure(figsize=(6,4))
avg_math_by_gender.plot(kind='bar', color=['skyblue','lightcoral'], edgecolor='black')
plt.title('Average Reading Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Reading Score')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['writing score'], bins=10, color='yellow', edgecolor='black')
plt.title('Distribution of Writing Scores')
plt.xlabel('Writing Score')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```
